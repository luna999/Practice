```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("Books dataset.csv")
```

```python
data.head()
```

```python
# Histogram for Price
# specify the number of bins, using 'bins' parameter
plt.hist(data['Price'], bins= 10)

# add the graph title and axes labels
plt.title('Distribution of price')
plt.xlabel('Price')
plt.ylabel('Frequency')

# display the plot
plt.show()
```

```python
# Count plot for Type
sns.countplot(data = data, x = 'Type')  #parameter hue for colour coding if there's another dimension

# display the plot
plt.show()
```

```python
# strip plot of columns 'Rating' and 'Type'
sns.stripplot(y = 'Rating', x = 'Type', data = data, jitter = True)  #jitter will avoid some overlaps and will shift some dots to left or right a bit to make the viz clearer

# display the plot
plt.show()
```

```python
# boxplot for rating
plt.boxplot(data['Rating'])

# add labels for five number summary
plt.text(x = 1.1, y = data['Rating'].min(), s ='min')
plt.text(x = 1.1, y = data['Rating'].quantile(0.25), s ='Q1')
plt.text(x = 1.1, y = data['Rating'].median(), s ='meadian (Q2)')
plt.text(x = 1.1, y = data['Rating'].quantile(0.75), s ='Q3')
plt.text(x = 1.1, y = data['Rating'].max(), s ='max')

# add the graph title and axes labels
plt.title('Boxplot of Rating')
plt.ylabel('Rating')

# display the plot
plt.show()
```

```python
# distribution plot of the column 'Reviews'
sns.displot(data['Reviews'], kind='kde')  
#kde = kernel density estimation, a way to estimate the probability density function of a continuous random variable

# display the plot
plt.show()
```

```python
# compute correlation
corr_matrix = data.corr()

corr_matrix
```

```python
# plot heatmap
# 'annot=True' returns the correlation values 
sns.heatmap(corr_matrix, annot = True)

# display the plot
plt.show()
```

```python
# Scatter plot of Number_of_pages and Price

# data
X = data['Number_Of_Pages']
Y = data['Price']

# plot the scatter plot
plt.scatter(X,Y)

# add the axes labels to the plot
plt.xlabel('Number_Of_Pages')
plt.ylabel('Price')

# display the plot
plt.show()
```

```python
data['Type'].value_counts()
```

```python
Typelist = data['Type'].value_counts().index.tolist()
```

```python
# Pieplot of Type
# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot
plt.pie(data['Type'].value_counts(), autopct = '%.1f%%', radius = 2.0, labels = Typelist)
plt.legend(bbox_to_anchor = (2.0, 1.3))
# display the plot
plt.show()
```
